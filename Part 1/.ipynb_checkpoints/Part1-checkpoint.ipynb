```python
#packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import seaborn as sns
import pandas as pd
```

```python
#parameters
x0 = 0
N = 10000
s = 1

#pdf given
def pdf(x):
    return 0.5 * np.exp(-np.abs(x))

#random walk metropolis given
def metropolis(x0, N, s):
    x = np.zeros(N)
    x[0] = x0

    for i in range(1, N):
        random_number = np.random.normal(loc=x[i-1], scale=s)
        r = pdf(random_number) / pdf(x[i-1])
        
        if np.log(np.random.uniform()) < np.log(r):
            x[i] = random_number
        else:
            x[i] = x[i-1]
    
    return x

#get samples and kernel density
samples = metropolis(x0, N, s)
kernel_density = gaussian_kde(samples)

#plots
x_vals = np.linspace(-5, 5, 1000)

plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, color="#818aa3", edgecolor='white', alpha=0.8, label="Samples Density Distribution")
plt.plot(x_vals, kernel_density(x_vals), color="#6d2e46", linewidth=2, label="Kernel Density Line")
plt.plot(x_vals, pdf(x_vals), color="#6adc8e", linewidth=2, label="Probability Density Function")
plt.xlim(-5, 5)
plt.ylim(0.0, 0.6)
plt.xticks(np.arange(-5, 6, 1))
plt.title("Random Walk Metropolis Samples")
plt.xlabel("x")
plt.legend(loc="upper right")
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()

# Monte Carlo estimates
mean_estimate = np.mean(samples)
sd_estimate = np.std(samples)

print("Monte Carlo Mean Estimate:", mean_estimate)
print("Monte Carlo Standard Deviation Estimate:", sd_estimate)
```

Exercise B

```python
#function to calculate Rhat

def calculate_r_hat(chains):
    J = len(chains)
    N = len(chains[0])

    means = np.array([np.mean(chain) for chain in chains])
    variances = np.array([np.var(chain, ddof=1) for chain in chains])

    W = np.mean(variances)
    M = np.mean(means)
    B = np.mean((means - M)**2)

    R_hat = np.sqrt((B + W) / W)
    return R_hat

#values of N and s given
N = 2000
s = 0.001
x0_initial_values = [0, 1, 2, 3]  # 4 chains with different initial values


chains = [metropolis(x0, N, s) for x0 in x0_initial_values]
R_hat = calculate_r_hat(chains)
print(f"R-hat for s = 0.001: {R_hat:.4f}")

#create plot for values of Rhat over grid of s values
s_values = np.arange(0.001, 1.001, 0.001)
R_hats = []

for s_val in s_values:
    chains = [metropolis(x0, N, s_val) for x0 in x0_initial_values]
    R_hats.append(calculate_r_hat(chains))


r_hat_df = pd.DataFrame({
    's': s_values,
    'R_hat': R_hats
})

r_hat_df['threshold_group'] = np.where(r_hat_df['R_hat'] > 1.05, "Above 1.05", "Below or 1.05")
r_hat_df.head(10)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=r_hat_df, x='s', y='R_hat', hue='threshold_group', palette={"Above 1.05": "#818aa3", "Below or 1.05": "#1b9e77"}, s=10)
plt.title("Values of R̂ over s values")
plt.xlabel("s")
plt.ylabel("R̂")
plt.grid(True, alpha=0.3)
plt.legend(title="R̂ vs Threshold")
plt.tight_layout()
plt.show()

#plot: zoomed ylim(0,1
plt.figure(figsize=(10, 6))
sns.scatterplot(data=r_hat_df, x='s', y='R_hat', hue='threshold_group', palette={"Above 1.05": "#818aa3", "Below or 1.05": "#1b9e77"}, s=10)
plt.title("Values of R̂ over s values (Zoomed In)")
plt.xlabel("s")
plt.ylabel("R̂")
plt.ylim(0, 2)
plt.grid(True, alpha=0.3)
plt.legend(title="R̂ vs Threshold")
plt.tight_layout()
plt.show()
```

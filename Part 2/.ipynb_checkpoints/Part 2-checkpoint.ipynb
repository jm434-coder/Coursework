```python
#libraries
import os
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
import seaborn as sns
from scipy import stats
from statsmodels.formula.api import ols
```

```python
#creating DB
conn = sqlite3.connect('airlines.db')

#check if airline already existed
try:
    pd.read_sql("SELECT 1 FROM ontime LIMIT 1;", conn) #by checking ontime e.g.
    print("Table 'ontime' exists.") #text just to confirm
except:
    print("Table not found: loading them") #text just to confirm

    # load csv
    years = [2004, 2005, 2006, 2007, 2008]
    ontime_list = [pd.read_csv(f"{year}.csv").rename(str.lower, axis='columns') for year in years]
    ontime = pd.concat(ontime_list, ignore_index=True)
    
    airports = pd.read_csv("airports.csv").rename(str.lower, axis='columns')
    carriers = pd.read_csv("carriers.csv").rename(str.lower, axis='columns')
    planes   = pd.read_csv("plane-data.csv").rename(str.lower, axis='columns')
    
    # save tables in db
    ontime.to_sql("ontime", conn, index=False, chunksize = 100000)
    airports.to_sql("airports", conn, index=False, chunksize = 100000)
    carriers.to_sql("carriers", conn, index=False, chunksize = 100000)
    planes.to_sql("planes", conn, index=False, chunksize = 100000)

#def query to make it easier for future use
def query(sql):
    return pd.read_sql(sql, conn)
```

Exercise A

```python
#check number of missings in CRSDepTime and DepTime
crs_dep_missings = query("SELECT COUNT(*) as n_missing FROM ontime WHERE crsdeptime IS NULL")['n_missing'][0]
dep_missings = query("SELECT COUNT(*) as n_missing FROM ontime WHERE deptime IS NULL")['n_missing'][0] 

print(f"Number of Missings in CRSDepTime: {crs_dep_missings}")
print(f"Number of Missings in DepTime: {dep_missings}")

missing_arrdelay_info = query("""
SELECT cancelled, diverted, COUNT(*) as n
FROM ontime
WHERE arrdelay IS NULL
GROUP BY cancelled, diverted
""")

display(missing_arrdelay_info) #all missing arrdelay were cancelled or diverted

#get dayofweek per year
best_days = query("""
SELECT year, 
       CASE dayofweek
            WHEN 1 THEN 'Monday'
            WHEN 2 THEN 'Tuesday'
            WHEN 3 THEN 'Wednesday'
            WHEN 4 THEN 'Thursday'
            WHEN 5 THEN 'Friday'
            WHEN 6 THEN 'Saturday'
            WHEN 7 THEN 'Sunday'
        END AS weekday_label,
       AVG(arrdelay) AS mean_arrdelay
FROM ontime
WHERE arrdelay IS NOT NULL
GROUP BY year, weekday_label
""")
best_days = best_days.sort_values(["year", "mean_arrdelay"]).groupby("year").head(1)

#get hours per year
best_hours = query("""
SELECT year, 
       CAST(crsdeptime / 100 AS INT) AS hour,
       AVG(arrdelay) AS mean_arrdelay
FROM ontime
WHERE arrdelay IS NOT NULL AND crsdeptime IS NOT NULL
GROUP BY year, hour
""")
best_hours = best_hours.sort_values(["year", "mean_arrdelay"]).groupby("year").head(1)

#combining: best day AND hour per year
best_day_hour = query("""
SELECT year, 
       CASE dayofweek
            WHEN 1 THEN 'Monday'
            WHEN 2 THEN 'Tuesday'
            WHEN 3 THEN 'Wednesday'
            WHEN 4 THEN 'Thursday'
            WHEN 5 THEN 'Friday'
            WHEN 6 THEN 'Saturday'
            WHEN 7 THEN 'Sunday'
        END AS weekday_label,
       CAST(crsdeptime / 100 AS INT) AS hour,
       AVG(arrdelay) AS mean_arrdelay
FROM ontime
WHERE arrdelay IS NOT NULL AND crsdeptime IS NOT NULL
GROUP BY year, weekday_label, hour
""")
best_day_hour = best_day_hour.sort_values(["year", "mean_arrdelay"]).groupby("year").head(1)

#time periods
def get_time_period(hour):
    if hour >= 23 or hour < 6:
        return "Night Time"
    elif hour < 10:
        return "Early Morning"
    elif hour < 12:
        return "Late Morning"
    elif hour < 14:
        return "Early Afternoon"
    elif hour < 17:
        return "Afternoon"
    elif hour < 20:
        return "Evening"
    else:
        return "Late Evening"

ontime = query("SELECT year, crsdeptime, arrdelay FROM ontime WHERE crsdeptime IS NOT NULL")
ontime['hour'] = ontime['crsdeptime'] // 100
ontime['time_period'] = ontime['hour'].apply(get_time_period)
best_times = (
    ontime.dropna(subset=["arrdelay"])
    .groupby(['year', 'time_period'])['arrdelay']
    .mean()
    .reset_index(name='mean_arrdelay')
    .sort_values(['year', 'mean_arrdelay'])
    .groupby('year')
    .head(1)
)

ontime = query("SELECT year, crsdeptime, arrdelay, dayofweek FROM ontime WHERE crsdeptime IS NOT NULL AND dayofweek IS NOT NULL")
ontime['hour'] = ontime['crsdeptime'] // 100
ontime['time_period'] = ontime['hour'].apply(get_time_period)
weekday_map = {1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Saturday", 7: "Sunday"}
ontime['weekday_label'] = ontime['dayofweek'].map(weekday_map)

#combining: best day AND time period per year
best_day_time = (
    ontime.dropna(subset=["arrdelay"])
    .groupby(['year', 'weekday_label', 'time_period'])['arrdelay']
    .mean()
    .reset_index(name='mean_arrdelay')
    .sort_values(['year', 'mean_arrdelay'])
    .groupby('year')
    .head(1)
)

#everything in pretty tables :)
print("Best Day to Fly per Year to Minimise Delays")
print(tabulate(best_days, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))

print("Best Hour to Fly per Year to Minimise Delays")
print(tabulate(best_hours, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))

print("Best Day and Hour to Fly per Year to Minimise Delays")
print(tabulate(best_day_hour, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))

print("Best Time Period to Fly per Year to Minimise Delays")
print(tabulate(best_times, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))

print("Best Day and Time Period to Fly per Year to Minimise Delays")
print(tabulate(best_day_time, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))
```

Exercise B

```python
#getting ontime and planes
query_sql = """
SELECT o.year AS year,
       o.arrdelay,
       p.model,
       p.year AS year_y,
       o.tailnum
FROM ontime AS o
INNER JOIN planes AS p
    ON o.tailnum = p.tailnum
"""
flight_age_delay = pd.read_sql(query_sql, conn)

#check why year.y is character and not integer - some values must have text
non_numeric_years = flight_age_delay[~flight_age_delay['year_y'].astype(str).str.match(r'^\d{4}$')]
non_numeric_years['year_y'].dropna().unique()


#mutating to numeric, forces all other values to NA
flight_age_delay['year_y'] = pd.to_numeric(flight_age_delay['year_y'], errors='coerce')

#check percentage of NA in year.y
missing_summary = flight_age_delay['year_y'].isna().mean() * 100
print(f"{missing_summary:.2f}% year_y values are missing")

flight_age_delay['missing_year_y'] = np.where(flight_age_delay['year_y'].isna(), 'Missing', 'Present')


#does model influence?
model_missing = (
    flight_age_delay.groupby(['model', 'missing_year_y'])
    .size().unstack(fill_value=0)
    .assign(percent_missing=lambda df: df['Missing'] / (df['Missing'] + df['Present']) * 100)
)
model_missing.head()

#the model 1121 is which tailnum?
flight_age_delay.loc[flight_age_delay['model'] == '1121', 'tailnum'].unique()

#place 1966 as year.y in all of model 1121:
flight_age_delay.loc[
    (flight_age_delay['year_y'].isna()) & (flight_age_delay['model'] == '1121'), 'year_y'] = 1966

# update missing_year_y
flight_age_delay['missing_year_y'] = np.where(flight_age_delay['year_y'].isna(), 'Missing', 'Present')

#is there a difference in arrdelay?
summary_stats = (
    flight_age_delay
    .groupby('missing_year_y')['arrdelay']
    .agg(['count', lambda x: x.isna().sum(), 'mean', 'median', 'std'])
    .rename(columns={'<lambda_0>': 'missing_arrdelay'})
)
print(summary_stats)

ttest = stats.ttest_ind(
    flight_age_delay.loc[flight_age_delay['missing_year_y'] == 'Missing', 'arrdelay'].dropna(),
    flight_age_delay.loc[flight_age_delay['missing_year_y'] == 'Present', 'arrdelay'].dropna(),
    equal_var=False
)
print(ttest)

#drop empty year.y
flight_age_delay = flight_age_delay[~flight_age_delay['year_y'].isna()]

#create airplane age
flight_age_delay['airplane_age'] = flight_age_delay['year'] - flight_age_delay['year_y']

#drop empty arrdelay and year_y below 0 and airplane age below 0
flight_age_delay = flight_age_delay[~flight_age_delay['arrdelay'].isna()]
flight_age_delay = flight_age_delay[(flight_age_delay['year_y'] > 0) & (flight_age_delay['airplane_age'] > 0)]

# Combine into a data frame statistics updated
summary_table = pd.DataFrame({
    'Statistic': ['Min', 'Q1', 'Median', 'Mean', 'Q3', 'Max'],
    'Airplane Age': flight_age_delay['airplane_age'].describe()[['min', '25%', '50%', 'mean', '75%', 'max']].values,
    'Year of Manufacture': flight_age_delay['year_y'].describe()[['min', '25%', '50%', 'mean', '75%', 'max']].values,
    'Arrival Delay': flight_age_delay['arrdelay'].describe()[['min', '25%', '50%', 'mean', '75%', 'max']].values
})
summary_table.round(2)


#and see visually
def plot_hist_box(df, column, color, title, xlabel):
    fig, axs = plt.subplots(2, 1, figsize=(10, 6), gridspec_kw={'height_ratios': [3, 1]})
    sns.histplot(df[column], bins=30, kde=False, ax=axs[0], color=color)
    axs[0].set_title(f"Distribution of {title}")
    axs[0].set_xlabel(xlabel)
    axs[0].set_ylabel("Count")
    sns.boxplot(x=df[column], ax=axs[1], color=color)
    axs[1].set_xlabel(xlabel)
    axs[1].set_yticks([])
    plt.tight_layout()
    plt.show()

plot_hist_box(flight_age_delay, 'airplane_age', '#818aa3', 'Airplane Age', 'Airplane Age (years)')
plot_hist_box(flight_age_delay, 'arrdelay', '#1b9e77', 'Arrival Delay', 'Arrival Delay (minutes)')
plot_hist_box(flight_age_delay, 'year_y', '#6d2e46', 'Year of Manufacture', 'Year of Manufacture')

#10 biggest delays and 10 most early
flight_age_delay[['arrdelay', 'tailnum', 'model', 'airplane_age', 'year']].nlargest(10, 'arrdelay')
flight_age_delay[['arrdelay', 'tailnum', 'model', 'airplane_age', 'year']].nsmallest(10, 'arrdelay')

#remove arrival delay below -120 (2h)
flight_age_delay = flight_age_delay[flight_age_delay['arrdelay'] >= -120]

#plot scatterplot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=flight_age_delay, x='airplane_age', y='arrdelay', hue='year', palette='viridis', alpha=0.3, s=10)
plt.title("Arrival Delay vs Airplane Age by Year")
plt.xlabel("Airplane Age (years)")
plt.ylabel("Arrival Delay (minutes)")
plt.xlim(0, 50)
plt.ylim(-100, 300)
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#age threshold
q3 = flight_age_delay['airplane_age'].quantile(0.75)
flight_age_delay['age_group'] = np.where(flight_age_delay['airplane_age'] >= q3, 'Older', 'Newer')

#make t tests and df per year
ttest_results = []

for year, group in flight_age_delay.groupby('year'):
    older = group[group['age_group'] == 'Older']['arrdelay'].dropna()
    newer = group[group['age_group'] == 'Newer']['arrdelay'].dropna()
    if len(older) > 1 and len(newer) > 1:
        ttest = stats.ttest_ind(older, newer, equal_var=False)
        ttest_results.append({
            'year': year,
            'p_value': ttest.pvalue,
            'mean_older': older.mean(),
            'mean_newer': newer.mean()
        })

pd.DataFrame(ttest_results).sort_values('year')
```
